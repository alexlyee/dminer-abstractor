```python
file_path_1 = '/content/drive/MyDrive/P_Misc/Climate-Forecasting/GCM_bc/file1.nc4'
data1 = xr.open_dataset(file_path_1)
file_path_2 = '/content/drive/MyDrive/P_Misc/Climate-Forecasting/GCM_bc/file2.nc4'
data2 = xr.open_dataset(file_path_2)
```

```python
def extract_data(data, variable="Air Temperature", day=1):
  level = data.lev.values
  lat = data.lat.values
  lon = data.lon.values
  time = data.time.values
  if variable == "Air Temperature": variable_values =  data.ta[day-1,0,:,:] #day, level, lat, lon
  if variable == "Sea-level pressure": variable_values =  data.psl[day-1,:,:]
  return variable_values
def plot_data(data, scenerio = "Historical", variable="Air Temperature", day = 1, month="April", year = "1994"):
  variable_values = extract_data(data, variable=variable, day=day)
  longitude = np.linspace(-180, 180, 288)
  latitude = np.linspace(-90, 90, 145)
  longitude, latitude = np.meshgrid(longitude, latitude)
  plt.figure(figsize=(10,6))
  plt.contourf(longitude, latitude, variable_values, cmap='jet')
  plt.colorbar()  # Add a colorbar for reference
  plt.title(f'{scenerio} {variable} for {day} {month}, {year}')  # Set a title for the plot
  plt.xlabel('Longitude')  # Label the x-axis
  plt.ylabel('Latitude')  # Label the y-axis
  plt.show()  # Display the plot
```

```python
plot_data(data1, scenerio = "Historical", variable="Air Temperature", day = 1, month="April", year = "1994")
plot_data(data1, scenerio = "Historical", variable="Air Temperature", day = 30, month="April", year = "1994")
plot_data(data1, scenerio = "Historical", variable="Sea-level pressure", day = 1, month="April", year = "1994")
plot_data(data1, scenerio = "Historical", variable="Sea-level pressure", day = 30, month="April", year = "1994")
```

```python
plot_data(data2, scenerio = "SSP245", variable="Air Temperature", day = 1, month="August", year = "2094")
plot_data(data2, scenerio = "SSP245", variable="Air Temperature", day = 30, month="August", year = "2094")
plot_data(data2, scenerio = "SSP245", variable="Sea-level pressure", day = 1, month="August", year = "2094")
plot_data(data2, scenerio = "SSP245", variable="Sea-level pressure", day = 30, month="August", year = "2094")
```
